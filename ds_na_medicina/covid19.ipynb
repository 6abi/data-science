##  import e Leitura dos dados

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv', sep=',')
```

```python
confirmed.head(5)
```

```python
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv', sep=',')
```

```python
deaths.head(5)
```

```python
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv', sep=',')
```

```python
recovered.head(5)
```

##  análises iniciais
***

##  Total de casos contágio
***

```python
confirmed['Country/Region'].value_counts()
```

```python
confirmed_by_country = confirmed.groupby('Country/Region').sum()
confirmed_by_country
```

```python
confirmed_by_country.loc['Brazil'][2:].plot()
```

```python
plt.figure(figsize=(10,5))
confirmed_by_country.loc['Brazil'][-50:].plot()
plt.title('Casos confirmados no Brasil')
plt.xlabel('Data')
plt.ylabel('Total')
plt.grid(True)
plt.show()
```

##  Novos casos de contágio
***

```python
new_cases = confirmed_by_country.loc['Brazil'][2:].diff().dropna()
```

```python
new_cases
```

```python
plt.figure(figsize=(10,5))
plt.title('Casos novos confirmados no Brasil')
new_cases.plot()
```

```python
last_date = confirmed_by_country.columns[-1]
last_date
```

```python
confirmed_by_country[last_date]
```

```python
plt.figure(figsize=(10,5))
plt.title('Casos confirmados por País(no último dia registrado na base)')
confirmed_by_country[last_date].sort_values(ascending=False)[0:10].plot(kind='bar')
```

```python
def latested_by_country(data):
    return data.groupby('Country/Region').sum().iloc[:,-1]
```

```python
confirmed_by_country = latested_by_country(confirmed)
deaths_by_country = latested_by_country(deaths)
recovered_by_country = latested_by_country(recovered)
```

```python
print("\nPaís com maior nº de casos ")
print(confirmed_by_country.sort_values(ascending=False)[0:1])
print("\nPaís com maior nº de mortes ")
print(deaths_by_country.sort_values(ascending=False)[0:1])
print("\nPaís com maior nº de recuperados ")
print(recovered_by_country.sort_values(ascending=False)[0:10])
```

```python
informations = [confirmed_by_country, deaths_by_country, recovered_by_country]
combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head(10)
```

##  Taxa de letalidade para a data da análise
#### expressa em %
##### representa o x% de mortes dentre os contaminados
