```python
def latested_by_country(data):
    return data.groupby('Country/Region').sum().iloc[:,-200]
```

```python
confirmed_by_country = latested_by_country(confirmed)
deaths_by_country = latested_by_country(deaths)
recovered_by_country = latested_by_country(recovered)
```

```python
print("\nPaís com maior nº de casos ")
print(confirmed_by_country.sort_values(ascending=False)[0:1])
print("\nPaís com maior nº de mortes ")
print(deaths_by_country.sort_values(ascending=False)[0:1])
print("\nPaís com maior nº de recuperados ")
print(recovered_by_country.sort_values(ascending=False)[0:10])
```

```python
informations = [confirmed_by_country, deaths_by_country, recovered_by_country]
combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

##  Taxa de letalidade para a data da análise
#### expressa em %
##### representa o x% de mortes dentre os contaminados
***

```python
print(combined.sum())
sum_up =combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed'] * 100
print(letality_rate_1)
```

```python
letality_rate_2 = combined['deaths'].sum() / (combined['deaths'].sum() + combined['recovered'].sum()) * 100
letality_rate_2
```

##### Observer que, como não temos números de recuperados, ainda, infere-se erroniamente, uma taxa de letalidade de 100% !
***

```python
combined['letality_rate_1'] = combined['deaths'] / combined['confirmed'] * 100
combined['letality_rate_2'] = combined['deaths'] / (combined['deaths']+ combined['recovered']) * 100
combined.head(10)
```

```python
combined.sort_values("letality_rate_1", ascending=False).head(10)
```

##  Taxa de letalidade para a data da análise

```python
import seaborn as sns
sns.scatterplot(data = combined, x = 'confirmed', y='letality_rate_1')
```
